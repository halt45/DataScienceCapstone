## QUESTION 1

Question 1 required the student to gather postal code information from wikipedia in order to
  clean the data and prepare it for analyzing. '\n' and duplicate entries were removed 
  as to make the data more presentable. 

```python
!pip install folium
!pip install wikipedia

import pandas as pd 
import wikipedia as wp
from bs4 import BeautifulSoup
import requests
import io

html = wp.page("List of postal codes of Canada: M").html().encode("UTF-8")


df = pd.read_html(html, header = 0)[0]


df = df[df.Borough != 'Not assigned']


df = df[df.Borough!= 0]
df.reset_index(drop = True, inplace = True)
i = 0
for i in range(0,df.shape[0]):
    if df.iloc[i][2] == 'Not assigned':
        df.iloc[i][2] = df.iloc[i][1]
        i = i+1
```

```python
df = df.groupby(['Postal Code','Borough'])['Neighbourhood'].apply(', '.join).reset_index()
df
```

## QUESTION 3

Question 3 required students to provide a cluster map that details how the postal codes are 
  grouped together via the long and lat data. 

```python

%matplotlib inline
!pip install shapely
!pip install geopandas
from shapely.geometry import Point
import geopandas as gpd
import matplotlib.pyplot as plt


html = wp.page("List of postal codes of Canada: M").html().encode("UTF-8")

df = pd.read_html(html, header = 0)[0]


df = df[df.Borough != 'Not assigned']


df = df.groupby(['Postal Code', 'Borough'])['Neighbourhood'].apply(list).apply(lambda x:', '.join(x)).to_frame().reset_index()

for index, row in df.iterrows():
    if row['Neighbourhood'] == 'Not assigned':
        row['Neighbourhood'] = row['Borough']


url="http://cocl.us/Geospatial_data"
s=requests.get(url).content
c=pd.read_csv(io.StringIO(s.decode('utf-8')))


c.columns = ['Postal Code', 'Latitude', 'Longitude']
df = pd.merge(c, df, on='Postal Code')


df = df[['Postal Code', 'Borough', 'Neighbourhood', 'Latitude', 'Longitude']]


df['Coordinates'] = list(zip(df.Longitude, df.Latitude))
df['Coordinates'] = df['Coordinates'].apply(Point)
gdf = gpd.GeoDataFrame(df, geometry='Coordinates')


cities = gpd.read_file(gpd.datasets.get_path('naturalearth_cities'))
ax = cities[cities.name == "Toronto"].plot(
    color='white', edgecolor='black')


gdf.plot(ax=ax, color='red')
plt.ylabel('Latitude')
plt.xlabel('Longitude')
plt.title('Longitude vs Latitude for Postal Code Data')
plt.show()
```
